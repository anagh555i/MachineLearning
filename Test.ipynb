```python
import numpy as np 
import pandas as pd 

import seaborn as sb 
import matplotlib.pyplot as plt 

from sklearn.preprocessing import StandardScaler, LabelEncoder 

from sklearn.model_selection import train_test_split
```

```python
df = sb.load_dataset("iris") 
df.info()
```

```python
le = LabelEncoder() 
ss = StandardScaler() 

Y = le.fit_transform(df["species"].values) 
X = ss.fit_transform(df.drop("species",axis=1).values) 
```

```python
X_train,X_test, Y_train, Y_test = train_test_split(X,Y, stratify=Y, test_size=0.2, random_state=10)
```

```python
from sklearn.svm import SVC 

from sklearn.metrics import accuracy_score, precision_score, recall_score, confusion_matrix, f1_score, roc_curve, auc,roc_auc_score
```

```python
svm = SVC(kernel="rbf", C=0.1) 
model = svm.fit(X_train,Y_train) 
Y_prediction = model.predict(X_test) 

accuracy = accuracy_score(Y_test, Y_prediction) 
print("accuracy:",accuracy) 

cm = confusion_matrix(Y_test, Y_prediction) 
sb.heatmap(cm, annot=True, fmt="0.2f") 

# fpr, tpr, threshold = roc_curve(Y_test, Y_prediction) # do seperately for each

# sb.lineplot(x=fpr,y=tpr) 
# sb.lineplot(x=[0,1],y=[0,1],markers="-") 
# plot.show() 

area = roc_auc_score(Y_test.reshape(-1,1),Y_prediction.reshape(-1,1), multi_class="ovr") 
print("AUC:",area) 
```

```python
from sklearn.ensemble import RandomForestClassifier
```

```python
rf = RandomForestClassifier() 
model = rf.fit(X_train,Y_train ) 

Y_prediction = model.predict(X_test) 
accuracy = accuracy_score(Y_test,Y_prediction) 
print("accuracy:",accuracy)
```
