```python
import kagglehub

# Download latest version
path = kagglehub.dataset_download("adityamhaske/bank-marketing-dataset")

print("Path to dataset files:", path)
```

```python
import pandas as pd 
import numpy as np 

from sklearn.preprocessing import LabelEncoder 
from sklearn.preprocessing import StandardScaler 
```

```python
df = pd.read_csv("../../../../../../Users/anagh/.cache/kagglehub/datasets/adityamhaske/bank-marketing-dataset/versions/3/bank.csv",delimiter=";") 
print(df.info())
```

```python
df=df.replace("unknown",np.nan) 
df=df.drop(["poutcome","job","education","contact"],axis=1)
df.isnull().sum() 
```

```python
df.info() 
le = LabelEncoder() 
for f in df: 
    if(df[f].dtype==object): 
        df[f]= le.fit_transform(df[f]) 
df.info()
```

```python
X = df.drop("y",axis=1).values 
Y = df["y"].values.reshape(-1,1)
```

```python
print(X.shape) 
print(Y.shape)
```

```python
ss = StandardScaler() 
X = ss.fit_transform(X) 
```

```python
from tensorflow import keras 
```

```python
model = keras.Sequential()
```

```python
model.add(keras.layers.Input(shape=(X.shape[1],)))
model.add(keras.layers.Dense(10,activation="sigmoid")) 
model.add(keras.layers.Dense(32,activation="relu")) 
model.add(keras.layers.Dense(1,activation="sigmoid")) 
model.summary() 
model.compile(optimizer=keras.optimizers.Adam(learning_rate=0.001),metrics=["accuracy"],loss="binary_crossentropy")
```

```python
model.fit(
    X,
    Y,
    epochs=10,
    batch_size=32,
    validation_split=0.2, 
    verbose=2
)
```

```python
for x in X[0:10,:]: 
    print(x.shape)
    print(model.predict(x.reshape(1,-1))) 

model.evaluate(X,Y,verbose=1)
```
